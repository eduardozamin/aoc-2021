```python
def parseInput(text):
    return [[[int(y) for y in x.split(",")] for x in entry.split(" -> ")] for entry in text.strip().split("\n")]
lines_of_vents = parseInput(day05_input)
print("Number of lines loaded:", len(lines_of_vents))
```

```python
maxD = max(np.array(lines_of_vents).flatten())+1
seafloor = np.zeros(shape=(maxD, maxD), dtype=int)
print(seafloor)
print(lines_of_vents)
```

```python
def applyLines(sf, lines):
    all_steps = []
    for p1, p2 in lines:
        if p1[0] != p2[0] and p1[1] != p2[1]:
            continue  # skip non-straight lines
        steps = list(np.unique(np.linspace(p1, p2, num=max(sf.shape), dtype=int), axis=0))
        if steps:
            all_steps += steps
    for step in all_steps:
        x = step[1]
        y = step[0]
        sf[x][y] += 1
seafloor = np.zeros(shape=(maxD, maxD), dtype=int)
applyLines(seafloor, lines_of_vents)
seafloor
```

```python
solution_part1 = len([p for p in seafloor.flatten() if p >= 2])
print("Day 05 - Part 1")
print("Points where at least two lines overlap:", solution_part1)
```

```python
def applyLines2(sf, lines):
    all_steps = []
    for p1, p2 in lines:
        if p1[0] != p2[0] and p1[1] != p2[1]:
            if abs(p1[0] - p2[0]) != abs(p1[1] - p2[1]):
                print("Skip:", p1, p2)
                continue  # skip non-straight lines
            steps = list(np.unique(np.linspace(p1, p2, num=abs(p1[0] - p2[0])+1, dtype=int), axis=0))                
        else:
            steps = list(np.unique(np.linspace(p1, p2, num=max(sf.shape), dtype=int), axis=0))
        if steps:
            all_steps += steps
    for step in all_steps:
        x = step[1]
        y = step[0]
        sf[x][y] += 1
seafloor2 = np.zeros(shape=(maxD, maxD), dtype=int)
applyLines2(seafloor2, lines_of_vents)
seafloor2
```
